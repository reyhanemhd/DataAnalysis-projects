```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

stock_data = pd.read_csv('stock_data.csv')
stock_data
```

```python
stock_data.plot(x='Date', y='FB', figsize=(15,5), linewidth=3, label = 'Facebook Stock Price')
plt.ylabel('price [$]')
plt.legend(loc='upper left')
plt.grid()
```
